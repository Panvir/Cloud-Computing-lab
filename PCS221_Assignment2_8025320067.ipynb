Q3. Write a program in Python to group similar elements into matrix.

```python
#Input list
data = [2,3,1,4,2,5,2,1,3,6,5,2,7,4,5]

#To store unique values in order
unique_values = []

#To store grouped rows
matrix = []

# Loop through each number in the list
for num in data:
    #If number already exists, add it to that group
    if num in unique_values:
        index = unique_values.index(num)
        matrix[index].append(num)
    #If number is new, create a new group
    else:
        unique_values.append(num)
        matrix.append([num])

#Displaying result
print("Grouped Matrix:")
for row in matrix:
    print(row)
```

Q5. Write a program in Python to create an n x n square matrix, where all the sub-matrix has the sum of opposite corner elements as even.

```python
def create_even_corner_matrix(n):
    matrix = np.zeros((n, n), dtype=int)
    for i in range(n):
        for j in range(n):
            matrix[i][j] = 2 * (i + j)
    return matrix

n = 4
matrix = create_even_corner_matrix(n)
print("Matrix:\n", matrix)

# Verify opposite corners of full matrix
print("\nTop-left corner:", matrix[0][0])
print("Bottom-right corner:", matrix[n-1][n-1])
print("Sum:", matrix[0][0] + matrix[n-1][n-1], "-> Even?", (matrix[0][0] + matrix[n-1][n-1]) % 2 == 0)
```

Q6 Write a program in Python to perform row-wise element addition in tuple matrix

```python
#Creating a matrix of tuples
tuple_matrix = [
    (10, 20, 30, 40),
    (15, 25, 35, 45),
    (5,  15, 25, 35)
]

print("Original tuple matrix:")
for row in tuple_matrix:
    print(row)

#Row-wise element addition
row_sums = [sum(row) for row in tuple_matrix]

print("\nRow-wise sums:", row_sums)
```
